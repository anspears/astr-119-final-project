```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import astropy
import sep
from astropy.io import fits
from matplotlib import rcParams

rcParams['figure.figsize']=[10.,8.]
```

```python
data_1 = astropy.io.fits.getdata("hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits")
data_2 = astropy.io.fits.getdata("hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits")
data_3 = astropy.io.fits.getdata("hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits")
```

```python
#show the image
m, s = np.mean(data_1), np.std(data_1)
plt.imshow(data_1, interpolation='nearest', vmin=m-s, vmax=m+s, origin='lower')
plt.colorbar();
```

```python
print(data_1.shape)
print(data_2.shape)
print(data_3.shape)
```

```python
rgb_1 = np.zeros((data_1.shape[0],data_1.shape[1],3))
```

```python
print(rgb_1.shape)
print(rgb_2.shape)
print(rgb_3.shape)
```

```python
d_1plot = data_1.copy()
d_2plot = data_2.copy()
d_3plot = data_3.copy()


m = np.mean(d_1plot)
s = np.std(d_1plot)
d_1plot = np.clip(d_1plot,m-s,m+s)
rgb_1[:,:,0] = 1-(d_3plot-d_3plot.min())/(d_3plot.max()-d_3plot.min())
rgb_1[:,:,1] = 1-(d_2plot-d_2plot.min())/(d_2plot.max()-d_2plot.min())
rgb_1[:,:,2] = 1-(d_1plot-d_1plot.min())/(d_1plot.max()-d_1plot.min())
print(d_1plot.min(),d_1plot.max())
```

```python
plt.imshow(rgb_1)
```

```python
m = np.mean(d_2plot)
s = np.std(d_2plot)
d_2plot = np.clip(d_2plot,m-s,m+s)
rgb_2[:,:,0] = (d_2plot-d_2plot.min())/(d_2plot.max()-d_2plot.min())
print(d_2plot.min(),d_2plot.max())
plt.imshow(rgb_2)
```

```python
m = np.mean(d_3plot)
s = np.std(d_3plot)
d_3plot = np.clip(d_3plot,m-s,m+s)
rgb_3[:,:,2] = (d_3plot-d_3plot.min())/(d_3plot.max()-d_3plot.min())
print(d_3plot.min(),d_3plot.max())
plt.imshow(rgb_3)
```
